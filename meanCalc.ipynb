```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcMean(file):
    df = pd.read_csv(file)
    c = 0
    for sent in range(len(df)):
        c += (df.iloc[sent][0]*df.iloc[sent][1])
    return c/df['sentiCR'].sum()
```

```python
def calcFreq(file):
    df = pd.read_csv(file)
    temp = {}
    for _, row in df.iterrows():
        temp.update({row[0]: row[1]})
    return temp
```

```python
go_fileList = ['golang_Nov2019-Jan2020-CR.csv', 'golang_Feb2020-Apr2020-CR.csv', 'golang_May2020-Jul2020-CR.csv', 'golang_Aug2020-Sep2020-CR.csv']
go_lst = [calcMean(file) for file in go_fileList]
go_df = pd.DataFrame({'Nov2019-Jan2020': go_lst[0], 'Feb2020-Apr2020': go_lst[1], 'May2020-Jul2020': go_lst[2], 'Aug2020-Sep2020': go_lst[3]}, index=[0])
go_df.plot.bar(yticks= np.linspace(0, -0.5, 11)[:4], xlabel='(Negative:-1 - Neutral:0 - Positive:1)', ylabel='Average Sentiment Score', figsize=(12,6)).legend(loc='upper right', title='Month of the Year')
plt.title('GoLang: Nov 2019 - Sep 2020 Average Sentiment Distribution', fontsize= 14, fontweight='bold')
```
